```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, OneHotEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
train_data = pd.read_csv('train-data.csv')
train_data
```

```python
test_data = pd.read_csv('test-data.csv')
test_data
```

```python
train_data = train_data.drop(['Unnamed: 0', 'New_Price'], axis=1)
test_data = test_data.drop(['Unnamed: 0', 'New_Price'], axis=1)
     
```

```python
train_data['Flag'] = 1
test_data['Flag'] = 0
```

```python
merge_data = pd.concat([train_data,test_data])
```

```python
merge_data['Mileage'] = merge_data['Mileage'].str.extract('(\d+\.\d+|\d+)').astype(float)
merge_data['Engine'] = merge_data['Engine'].str.extract('(\d+)').astype(float).fillna(0).astype(int)
merge_data['Power'] = merge_data['Power'].str.extract('(\d+\.\d+|\d+)').astype(float)
```

```python
merge_data.to_csv("merge_data.csv")
merge_data
```

```python
label_encoder = LabelEncoder()
categorical_cols = ['Name', 'Location', 'Fuel_Type', 'Transmission', 'Owner_Type']
for col in categorical_cols:
    merge_data[col] = label_encoder.fit_transform(merge_data[col])

onehot_encoder = OneHotEncoder(sparse=False)
encoded_data = onehot_encoder.fit_transform(merge_data)
```

```python
X_train = encoded_data[merge_data['Flag'] == 1, :-1]
y_train = encoded_data[merge_data['Flag'] == 1, -1]
X_test = encoded_data[merge_data['Flag'] == 0, :-1]
X_train = np.nan_to_num(X_train)
y_train = np.nan_to_num(y_train)
X_test = np.nan_to_num(X_test)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
     
```

```python
model = Sequential()
model.add(Dense(128, activation='relu', input_dim=X_train_scaled.shape[1]))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='relu'))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train_scaled, y_train, epochs=15, batch_size=40)
predictions = model.predict(X_test_scaled)

print(predictions)
```
